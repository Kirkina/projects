```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0'}
```

```python
result_df = pd.DataFrame()
num_of_page = 3
for i in range(num_of_page):
    # получаем список ссылок на каждой странице
    url = 'https://nn.hh.ru/search/vacancy?L_is_autosearch=false&ored_clusters=true&employer_id=3529&page=' + str(i + 1)
    r = requests.get(url + str(i), headers=headers)
    content = r.text
    
    soup = BeautifulSoup(content, 'lxml')
    box = soup.find_all('a', class_='serp-item__title')
    links = [link['href'] for link in box]
    
    # переходим по каждой из ссылок
    for link in links:
        result = requests.get(link, headers=headers) # отправляем запрос
        content = result.text 
        soup = BeautifulSoup(content, 'lxml')

        box = soup.find('div', class_='vacancy-title')
        if box is not None: # Проверяем, что элемент существует
            title = box.find('h1').get_text()

            salary_element = box.find('span', class_='bloko-header-section-2_lite')
            if salary_element is not None: # Проверяем, что элемент существует
                salary_text = salary_element.text.strip() # Получаем текст из элемента
                salary = salary_text.split('до вычета налогов')[0].strip()
            else:
                salary = 'No information'

            new_row = {'Ссылка': link, 'Название вакансии': title, 'ЗП': salary}
            result_df = result_df.append(new_row, ignore_index=True)
            result_df.to_excel('r.xlsx')
```

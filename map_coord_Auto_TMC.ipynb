```python
from math import radians, cos, sin, asin, sqrt


def haversine(lat1, lon1, lat2, lon2):
    lon1, lat1, lon2, lat2 = map(radians, (lon1, lat1, lon2, lat2))
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * asin(sqrt(a))
    km = 6367 * c
    return km


def haversine2(lat1, lon1, lat2, lon2):
    coord_1 = (lat1, lon1)
    coord_2 = (lat2, lon2)
    res_km = geopy.distance.geodesic(coord_1, coord_2).km
    return res_km
        
print(haversine(58.129509, 52.603714, 58.140835, 52.670921))
print(haversine2(58.129509, 52.603714, 58.140835, 52.670921))
```

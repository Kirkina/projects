```python
kl = pd.read_excel('клиники_с_координатами.xlsx')
```

```python
kliniki = pd.merge(kliniki, kl, left_on = ['Наименование клиники', 'lat', 'long'], right_on = ['Наименование клиники', 'lat', 'long'], how = 'left') 
```

```python
len(kliniki)
kliniki.head()
```

```python
del kliniki['url_y']
del kliniki['url_with_coordinates_y']
```

```python
kliniki.head()
```

```python
sotr = pd.read_excel('сотрудники.xlsx')
```

```python
sotr.head()
```

```python
import pandas as pd
from scipy.spatial.distance import cdist

# Создаем матрицу расстояний между сотрудниками и клиниками
distances = cdist(sotr[['lat', 'long']], kliniki[['lat', 'long']], metric='euclidean')

# Находим индексы ближайших клиник для каждого сотрудника
closest_clinic_indices = distances.argmin(axis=1)

# Добавляем информацию о ближайших клиниках в таблицу с сотрудниками
sotr['closest_clinic'] = kliniki.loc[closest_clinic_indices, 'Наименование клиники'].values
sotr['clinic_address'] = kliniki.loc[closest_clinic_indices, 'address'].values
```

```python
sotr.head()
```

```python
sotr['distance_to_closest_clinic'] = distances.min(axis=1)
```

```python
sotr.to_excel('ближ_клиника.xlsx')
```
